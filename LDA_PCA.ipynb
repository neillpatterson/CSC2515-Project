```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.lda import LDA
from loadData import load
from mpl_toolkits.mplot3d import Axes3D

%matplotlib inline
```

# PCA

```python
trainData, trainLabels, testData, testLabels = load()
```

```python
pca = PCA(n_components=3)
#X = pca.inverse_transform(pca.fit_transform(trainData))
X = pca.fit_transform(trainData)
```

```python
fig = plt.figure(figsize=(7,7))
ax = Axes3D(fig)

idx1 = np.array(trainLabels==1).ravel()
idx2 = np.array(trainLabels==2).ravel()
idx3 = np.array(trainLabels==3).ravel()
idx4 = np.array(trainLabels==4).ravel()
idx5 = np.array(trainLabels==5).ravel()
idx6 = np.array(trainLabels==6).ravel()

ax.plot(X[idx1,0], X[idx1,1], X[idx1,2], 'bo', ms=3, label="walking")
ax.plot(X[idx2,0], X[idx2,1], X[idx2,2], 'ro', ms=3, label="walking upstairs")
ax.plot(X[idx3,0], X[idx3,1], X[idx3,2], 'go', ms=3, label="walking downstairs")
ax.plot(X[idx4,0], X[idx4,1], X[idx4,2], 'co', ms=3, label="sitting")
ax.plot(X[idx5,0], X[idx5,1], X[idx5,2], 'yo', ms=3, label="standing")
ax.plot(X[idx6,0], X[idx6,1], X[idx6,2], 'mo', ms=3, label="laying")

ax.set_xticklabels([])
ax.set_yticklabels([])
ax.set_zticklabels([])
ax.legend()
```

# LDA

```python
trainData, trainLabels, testData, testLabels = load()
```

```python
lda = LDA(solver="svd", store_covariance=True)
X = lda.fit_transform(trainData, trainLabels)
```

```python
import itertools
xyz =  list(itertools.combinations(np.arange(5),3))

fig = plt.figure(figsize=(24,48))

for i in range(10):
    x,y,z = xyz[i]
    ax = fig.add_subplot(5,2, i+1, projection='3d', axisbg='white')

    ax.set_xticklabels([])
    ax.set_yticklabels([])
    ax.set_zticklabels([])
    #ax.w_xaxis.set_pane_color((1.0,1.0,1.0,1.0))
    #ax.w_yaxis.set_pane_color((1.0,1.0,1.0,1.0))
    #ax.w_zaxis.set_pane_color((1.0,1.0,1.0,1.0))

    ax.plot(X[idx1,x], X[idx1,y], X[idx1,z], 'bo', ms=3, label="walking")
    ax.plot(X[idx2,x], X[idx2,y], X[idx2,z], 'ro', ms=3, label="walking upstairs")
    ax.plot(X[idx3,x], X[idx3,y], X[idx3,z], 'go', ms=3, label="walking downstairs")
    ax.plot(X[idx4,x], X[idx4,y], X[idx4,z], 'co', ms=3, label="sitting")
    ax.plot(X[idx5,x], X[idx5,y], X[idx5,z], 'yo', ms=3, label="standing")
    ax.plot(X[idx6,x], X[idx6,y], X[idx6,z], 'mo', ms=3, label="laying")
    
    if (i % 2 == 1):
        ax.legend(loc='center left', bbox_to_anchor=(1,0.5), fontsize=30, markerscale=8)

fig.tight_layout(pad=1.0, w_pad=-15, h_pad=-4)
```
